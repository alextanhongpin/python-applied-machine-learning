```python
import pandas as pd

df = pd.read_csv("reviews.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
df.groupby("Sentiment").describe()
```

```python
df.drop_duplicates(inplace=True)
df.groupby("Sentiment").describe()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1, 2), stop_words="english", min_df=20)

X = vectorizer.fit_transform(df["Text"])
y = df["Sentiment"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay as cmd

cmd.from_estimator(
    model,
    X_test,
    y_test,
    display_labels=["Negative", "Positive"],
    cmap="Blues",
    xticks_rotation="vertical",
);
```

```python
text = 'The long lines and poor customer service really turned me off'
model.predict_proba(vectorizer.transform([text]))[0][1] # probablity positive
```

```python
text = 'The food was great and the service was excellent'
model.predict_proba(vectorizer.transform([text]))[0][1] # probablity positive
```
